```python
# Task 2: String Manipulation

sample_string = "I am trying to learn Python!"
```

```python
print(sample_string)
print(sample_string.find("t"))    # showing how many times 't' is present in my string
print(sample_string.islower())    # showing the case of my string
print(sample_string.replace("I am", "I'm"))    # replacing words within my string
```

```python
# Task 3: Basic Operations

a, b, c = 3, 5, 7
```

```python
print(a+b+c, a-b-c, a*b, c/a, a%b, a**c, b//c)    # Arithmetic Operations
print(a==b, b!=c, c>a, c<a, b>=a, b<=c)    # Comparison Operations
print(c>20 and b==5, c<5 or c>5, not a==3)    # Logical Opterations
```
